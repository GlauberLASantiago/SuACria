# SuACria - Super Auxiliar de Criatividade
Chatbot que fornece inspiração para ideias situações que necessitem de abordagens criativas. Tem como base a IA Gemini, do Google.
Autoria: Glauber Santiago (2024)

```python
## A seguir, temos as configurações iniciais da API:

# Instalando o SDK (Software Development Kit) do Python para a API Gemini.
!pip install -q -U google-generativeai

# Importando o modelo generativo.
import google.generativeai as genai

# Deixando a API key oculta, dentro dos dados do desenvolvedor.
from google.colab import userdata
GOOGLE_API_KEY=userdata.get('MINHA_API_KEY')
genai.configure(api_key=GOOGLE_API_KEY)

#Iniciando o Modelo generativo
model = genai.GenerativeModel('gemini-1.0-pro-latest')


## A seguir, temos as configurações iniciais do modelo:

generation_config = {
    "candidate_count": 1,
    "temperature": 0.7,
    "top_k": 40,
    "top_p": 0.95,
    "typical_p": 1,
    "num_samples": 1,
}


## Aprimoramento visual:
import textwrap
from IPython.display import display
from IPython.display import Markdown
def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))


## A seguir, temos as instruções para o chatbot em si:
chat = model.start_chat(history=[])
prompt = "Atue como um super auxiliar de criatividade. Ou seja, o usuário é um artista, e você irá ajudá-lo a encontrar inspiração. Para isso, você deverá perguntar sobre qual tema você o usuário necessita de sua ajuda:"
response = chat.send_message(prompt)
print("\n", "❀ SuACria:", response.text, "\n")
prompt = input("❄ Você: ")


while prompt != "Fechar":
    response = chat.send_message(prompt)
    print("\n", "❀ SuACria:", response.text, "\n", "\n", "Continue a conversa ou digite 'Fechar' para terminar.", "\n", "\n")
    prompt = input("❄ Você: ")

print("\n❀ SuACria: Agradeço por conversar comigo! Até a próxima!")
```
